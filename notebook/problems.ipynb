## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# TODO
from scipy.stats import norm

def norm_density(x, mean=0, sd= 1):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = norm.pdf(x, loc= mean, scale= sd)

    return result

def norm_cumulative(q, mean=0, sd= 1):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = norm.cdf(q, loc= mean, scale= sd)

    return result

def norm_quantile(p, mean=0, sd= 1):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = norm.ppf(p, loc= mean, scale= sd)

    return result

def norm_random(n, mean=0, sd= 1):
    """
    Generates random variables from the binomial distribution
    """
    result = norm.rvs(size= n, loc= mean, scale= sd)

    return result
```

```python
np.random.seed(42)

mean = 0
sd = 1
data = norm_random(2000, mean = mean, sd = sd)

pdf = norm_density(data, mean = mean, sd = sd)
cdf = norm_cumulative(data, mean = mean, sd = sd)
ppf = norm_quantile(np.linspace(0, 1, len(data)), mean = mean, sd = sd)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pdf, "r-", basefmt=" ")
axis[0].set_title('PDF of Normanl Distribution')
axis[1].stem(data, cdf, "b-", basefmt=" ")
axis[1].set_title('CDF of Normanl Distribution')
axis[2].stem(np.linspace(0, 1, len(data)), ppf, "g-", basefmt=" ")
axis[2].set_title('PPF of Normanl Distribution')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import uniform

def uniform_density(x, min=0, max= 1):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = uniform.pdf(x, loc= min, scale= (max-min))

    return result

def uniform_cumulative(q, min=0, max= 1):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = uniform.cdf(q, loc= min, scale= (max-min))

    return result

def uniform_quantile(p, min=0, max= 1):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = uniform.ppf(p, loc= min, scale= (max-min))

    return result

def uniform_random(n, min=0, max= 1):
    """
    Generates random variables from the binomial distribution
    """
    result = uniform.rvs(size= n, loc= min, scale= (max-min))

    return result
```

```python
np.random.seed(42)

min_value = 0
max_value = 1
data = uniform_random(2000, min = min_value, max = max_value)

pdf = uniform_density(data, min = min_value, max = max_value)
cdf = uniform_cumulative(data, min = min_value, max = max_value)
ppf = uniform_quantile(np.linspace(0, 1, len(data)), min = min_value, max = max_value)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pdf, "r-", basefmt=" ")
axis[0].set_title('PDF of Unifrom Distribution')
axis[1].stem(data, cdf, "b-", basefmt=" ")
axis[1].set_title('CDF of Uniform Distribution')
axis[2].stem(np.linspace(0, 1, len(data)), ppf, "g-", basefmt=" ")
axis[2].set_title('PPF of Uniform Distribution')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
